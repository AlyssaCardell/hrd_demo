### If-Else Conditions
In C#, if-else conditions are the primary way to control the flow of your program by making decisions based on whether a specific condition is true or false.

+ **if** statement: Executes a block of code only if the specified condition is true
+ **else** statement: Specifies a block of code to run if every condition is false
+ **else if** statement: Tests a new condition if the previous condition was false. You can chain multiple else-if statements together
+ **switch** statement: An alternative for testing a single variable against many potential values, often cleaner than multiple else-if blocks

````polyglot-notebook
using System;

class Program
{
    static void Main()
    {
        // basic if else condition
        int temperature = 25;

        // syntax
        /*
            if (condition goes here) {
                output
            }

            else {
                output
            }
        */

        if (temperature > 30) {
            Console.WriteLine("Its hot outside");
        } else {
            Console.WriteLine("Its a nice day");
        }

    }
}
````

### Comparison Operators 

+ **==** : Equal to
+ **!=** : Not equal to
+ **> / <** : Greater than / Less than
+ **>= / <=** : Greater than or equal to / Less than or equal to
+ **&& / ||** : Logical AND / Logical OR (used to combine conditions)

````polyglot-notebook
using System;

class Program
{
    static void Main()
    {
        // more than two conditions
        int score = 85;
        string name = "Aly";

        // syntax
        /*
            if (condition goes here) {
                output
            }

            else if (another condition) {
                a different output
            }

            else {
                if nothing else is true, this returns
            }
        */

        if (score >= 90) {
            Console.WriteLine("Grade: A");
        } else if (score >= 80) {
            Console.WriteLine("Grade: B");
        } else {
            Console.WriteLine("Grade: C or below");
        }

        // Logical AND & OR
        if (score == 85 && name == "Aly") {
            Console.WriteLine("Aly got 85");
        } else if (score == 90 || name == "Aly") {
            Console.WriteLine("This is just Aly");
        } else if (score == 90 && name == "James") {
            Console.WriteLine("This isn't even Aly anymore");
        }
    }
}
````
