## Console.ReadLine()        

What is Console.ReadLine()?
Console.ReadLine() is a method in the System namespace that:

Reads a line of text from the console (standard input).
Waits for the user to type something and press Enter.
Returns the entered text as a string.
Automatically removes the newline character (\n) at the end.

It is the most common way to get text input from the user in a C# console application.
Important Points:

It returns string? (nullable string in C# 8+).
If the user just presses Enter without typing anything, it returns an empty string "".
If the input stream ends (e.g. Ctrl+Z in Windows or Ctrl+D in Linux), it returns null.


using System;

class Program


{
    static void Main()
    
    {
        Console.Write("Enter your name: ");
        string name = Console.ReadLine();           // Read user input

        Console.Write("Enter your age: ");
        string ageInput = Console.ReadLine();

        // Convert string to int (with basic error handling)
        int age = int.TryParse(ageInput, out int result) ? result : 0;

        Console.WriteLine($"\nHello, {name}!");
        Console.WriteLine($"You are {age} years old.");
    }
}

### If-Else Conditions
In C#, if-else conditions are the primary way to control the flow of your program by making decisions based on whether a specific condition is true or false.

+ **if** statement: Executes a block of code only if the specified condition is true
+ **else** statement: Specifies a block of code to run if every condition is false
+ **else if** statement: Tests a new condition if the previous condition was false. You can chain multiple else-if statements together
+ **switch** statement: An alternative for testing a single variable against many potential values, often cleaner than multiple else-if blocks

````polyglot-notebook
using System;

class Program
{
    static void Main()
    {
        // basic if else condition
        int temperature = 25;

        // syntax
        /*
            if (condition goes here) {
                output
            }

            else {
                output
            }


            // without spacing
            if (){} else (){}
        */

        if (temperature > 30) {
            Console.WriteLine("Its hot outside");
        } else {
            Console.WriteLine("Its a nice day");
        }

    }
}
````

### Comparison Operators 

+ **==** : Equal to
+ **!=** : Not equal to
+ **> / <** : Greater than / Less than
+ **>= / <=** : Greater than or equal to / Less than or equal to
+ **&& / ||** : Logical AND / Logical OR (used to combine conditions)

````polyglot-notebook
using System;

class Program
{
    static void Main()
    {
        // more than two conditions
        int score = 85;
        string name = "Aly";

        // syntax
        /*
            if (condition goes here) {
                output
            }

            else if (another condition) {
                a different output
            }

            else {
                if nothing else is true, this returns
            }
        */

        if (score <= 90) {
            Console.WriteLine("Grade: A");
        } else if (score >= 80) {
            Console.WriteLine("Grade: B");
        } else {
            Console.WriteLine("Grade: C or below");
        }

        // Logical AND & OR
        if (score == 85 && name == "Aly") {
            Console.WriteLine("Aly got 85");
        } else if (score == 90 || name == "Aly") {
            Console.WriteLine("This is just Aly");
        } else if (score == 90 && name == "James") {
            Console.WriteLine("This isn't even Aly anymore");
        }
    }
}
````
